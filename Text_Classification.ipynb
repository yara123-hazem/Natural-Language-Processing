# Naive Bayes Implementaion


> Generating The Vocabulary from Corpus



```python
# Combine all text data into a single string
all_text = ' '.join(df_train['sentence'])

# Split the text into lowercase words
words = all_text.lower().split()

# Create a set to store unique words (vocab)
vocab = set(words)

# Remove non alphanumeric characters
vocab = {word for word in vocab if any(char.isalnum() for char in word)}

# Use set comprehension to split strings and add the split parts to the set
delimiter = ","
vocab = {part.strip() for item in vocab for part in item.split(delimiter)}

delimiter = "-"
vocab = {part.strip() for item in vocab for part in item.split(delimiter)}

delimiter = "\\/"
vocab = {part.strip() for item in vocab for part in item.split(delimiter)}

delimiter = "."
vocab = {part.strip() for item in vocab for part in item.split(delimiter)}

delimiter = "'"
vocab = {part.strip() for item in vocab for part in item.split(delimiter)}

delimiter = "&"
vocab = {part.strip() for item in vocab for part in item.split(delimiter)}

# Remove single character strings
vocab = {word for word in vocab if len(word) > 1}

vocab = vocab - {""}
```



> Implementation From Scratch



```python
import numpy as np

class NaiveBayes:
    def train_naive_bayes(self, D, C, vocab):
        # Calculate P(c) terms
        Ndoc = len(D) # Number of documents in the dataset D
        logprior = {}  # A dictionary to store the log prior probabilities for each class p(c)
        loglikelihood = {}  # A nested dictionary to store the log likelihoods of each word given each class p(x/c)
        Nc = {}   # A dictionary to store the count of documents for each class
        bigdoc = {}  # A dictionary to store documents grouped by class

        # Calculate Nc
        for c in C:
            Nc[c] = 0
            bigdoc[c] = []

        for doc, cls in D:
            Nc[cls] += 1
            bigdoc[cls].append(doc)

        # Calculate log P(c) terms
        for c in C:
            logprior[c] = np.log(Nc[c] / Ndoc)

        # Calculate log P(w|c) terms
        for w in vocab:
            for c in C:
                count_w_c = sum(1 for doc in bigdoc[c] if w in doc.lower().split())
                if w not in loglikelihood:
                    loglikelihood[w] = {}
                # Calculate the loglikelihood using Laplace Smoothing
                loglikelihood[w][c] = np.log((count_w_c + 1) / (len(vocab) + sum(len(doc.lower().split()) for doc in bigdoc[c])))

        return logprior, loglikelihood

    def pred(self, X, logprior, loglikelihood, vocab, C):
        predictions = []
        for x in X:
            posteriors = []
            for c in C:
                posterior = logprior[c]
                for word in x.lower().split():
                    if word in vocab:
                        posterior += loglikelihood[word][c]
                posteriors.append(posterior)
            predicted_class = C[np.argmax(posteriors)]
            predictions.append(predicted_class)

        return predictions

# Convert DataFrame to the appropriate format
D = list(zip(df_train.iloc[:, 0], df_train.iloc[:, 1]))
C = np.array([0, 1, 2, 3, 4])
nb=NaiveBayes()
logprior, loglikelihood =nb.train_naive_bayes(D, C, vocab)
```

```python
# Predict labels of test dataset
test=df_test['sentence']
predict=nb.pred(test,logprior,loglikelihood, vocab,C)
print("Predicted Labels:")
predict[0:10]
```

```python
import numpy as np

def test_naive_bayes(testdoc, logprior, loglikelihood, C, V):
    best_c = None
    best_sum = float('-inf')

    for c in C:
        sum_c = logprior[c]
        for s in testdoc['sentence']:  # Iterating directly over the elements of testdoc
            words = s.lower().split()
            for word in words:
              if word in V:
                  sum_c += loglikelihood.get(word, {}).get(c, 0)  # Adding loglikelihood if word exists in the vocabulary

        if sum_c > best_sum:
            best_sum = sum_c
            best_c = c

    return best_c

# Example usage:
testdoc = df_test
best_c = test_naive_bayes(testdoc, logprior, loglikelihood, C, vocab)
print("Best class:", best_c)
```



> Implementation With Scikit Learn



```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline

# Create a pipeline with CountVectorizer and MultinomialNB
pipeline = Pipeline([
    ('vectorizer', CountVectorizer(binary=False)), # Generate Vocab
    ('classifier', MultinomialNB(alpha=1.0)),  # Laplace smoothing
])

# Fit the pipeline to the training data
pipeline.fit(df_train.iloc[:, 0], df_train.iloc[:, 1])

# Extract vocabulary from CountVectorizer
vectorizer = pipeline.named_steps['vectorizer']
vocabulary = vectorizer.get_feature_names_out()

# Calculate log prior probabilities (log P(c))
Ndoc = len(df_train)
logprior2 = {}
for c in pipeline.classes_:
    Nc = (df_train.iloc[:, 1] == c).sum()
    logprior2[c] = np.log(Nc / Ndoc)

# Calculate log likelihoods (log P(w|c))
loglikelihood2 = {}
for c, class_index in enumerate(pipeline.classes_):
    class_count = pipeline.named_steps['classifier'].feature_count_[c]
    loglikelihood2[c] = {}
    for word_index, word in enumerate(vocabulary):
        word_count = class_count[word_index]
        loglikelihood2[c][word] = np.log((word_count + 1) / (class_count.sum() + len(vocabulary)))
```


# Confusion Matrix & Evaluation Metrics For Naive Bayes



> From Scratch



```python
def confusion_matrix(y_true, y_pred, num_classes):
    matrix = np.zeros((num_classes, num_classes), dtype=int)
    for i in range(len(y_true)):
        matrix[y_true[i], y_pred[i]] += 1
    return matrix

def precision_recall_f1(conf_matrix):
    num_classes = conf_matrix.shape[0]
    precision = np.zeros(num_classes)
    recall = np.zeros(num_classes)
    f1_score = np.zeros(num_classes)
    for i in range(num_classes):
        true_positive = conf_matrix[i, i]
        false_positive = np.sum(conf_matrix[:, i]) - true_positive
        false_negative = np.sum(conf_matrix[i, :]) - true_positive

        precision[i] = true_positive / (true_positive + false_positive) if (true_positive + false_positive) > 0 else 0
        recall[i] = true_positive / (true_positive + false_negative) if (true_positive + false_negative) > 0 else 0

        f1_score[i] = 2 * (precision[i] * recall[i]) / (precision[i] + recall[i]) if (precision[i] + recall[i]) > 0 else 0

    macro_precision = np.mean(precision)
    macro_recall = np.mean(recall)
    macro_f1_score = np.mean(f1_score)

    return precision, recall, f1_score, macro_precision, macro_recall, macro_f1_score
```

```python
import numpy as np

# Function to convert a sentence to vector representation
def sentence_to_vector(sentence, bi_gram_to_index):
    words = ['<S>'] + sentence.lower().split() + ['<E>'] # Convert to lowercase and split into words
    bi_grams = list(zip(words, words[1:]))  # Create bi-grams
    vector = np.zeros(len(bi_gram_to_index), dtype=np.int16)  # Initialize the vector with zeros

    for bi_gram in bi_grams:
        if bi_gram in bi_gram_to_index:
            vector[bi_gram_to_index[bi_gram]] = 1  # Set the corresponding index to 1
    return vector
```

```python
# Convert each sentence to a vector representation
feature_vectors = [sentence_to_vector(s, stoi) for s in df_train['sentence']]

# Convert the list of feature vectors to a NumPy array
feature_matrix_np = np.array(feature_vectors)

# Display the feature matrix
print("Feature Matrix:")
print(feature_matrix_np)
```


> Implementation From Scratch



```python
import numpy as np

class LogisticRegression():
    def __init__(self, n_iters=100, learning_rate=0.01):
        self.n_iters = n_iters
        self.learning_rate = learning_rate
        self.weights = None
        self.bias = None

    def softmax(self, x):
        exp_x = np.exp(x - np.max(x))  # Subtract the maximum value for numerical stability
        return exp_x / np.sum(exp_x, axis=1, keepdims=True)

    def compute_loss(self, y_true, y_pred):
        # Categorical cross-entropy
        epsilon = 1e-9
        num_samples = len(y_true)
        loss = -np.sum(y_true * np.log(y_pred + epsilon)) / num_samples
        del epsilon, num_samples
        return loss

    def feed_forward(self, X):
        z = np.dot(X, self.weights) + self.bias
        A = self.softmax(z)
        del z
        return A

    def fit(self, X, y):
        n_samples, n_features = X.shape

        # init parameters
        self.weights = np.zeros((n_features, y.shape[1]))
        self.bias = np.zeros((1, y.shape[1]))

        # SGD optimization
        for _ in range(self.n_iters):
            print("iter:", _)
            shuffled_indices = np.random.permutation(n_samples)
            X_shuffled = X[shuffled_indices]
            y_shuffled = y[shuffled_indices]

            for i in range(n_samples):
                X_i = X_shuffled[[i], :]
                y_i = y_shuffled[[i], :]
                y_pred = self.feed_forward(X_i)

                # Compute gradients
                # Assuming Categorical Cross-Entropy Loss
                dz = y_pred - y_i   # dL/dz
                dw = np.dot(X_i.T, dz)  # dL/dw
                db = np.sum(dz, axis=0, keepdims=True) # dL/db

                # Update parameters
                self.weights -= self.learning_rate * dw
                self.bias -= self.learning_rate * db

            del shuffled_indices, X_shuffled, y_shuffled

    def predict(self, X):
        z = np.dot(X, self.weights) + self.bias
        y_predicted = self.softmax(z)
        y_predicted_cls = np.argmax(y_predicted, axis=1)
        return y_predicted_cls


num_classes = 5
input_size = feature_matrix_np.shape[1]

model = LogisticRegression(n_iters=100, learning_rate=0.01)
labels = df_train['label'].to_numpy()
labels = np.eye(num_classes)[labels]  # One-hot encode labels
model.fit(feature_matrix_np, labels)


predictions = model.predict(test_matrix_np)
```
